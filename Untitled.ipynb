```python
import os
import rasterio as rio
import matplotlib.pyplot as plt
import numpy as np

# define path to dem
teton_folder = os.path.join('N:',os.sep,'RStor','dmcgrath','Glacier','Teton Project')
dem_path = os.path.join(teton_folder, 'DEMs', 'USGS DEMs', 'USGS_DEM_1m.tif')

# read in dem with rasterio
dem = rio.open(dem_path).read()

# set the nan value
dem[dem<0] = np.nan

# plot the dem with matplotlib
fig,axs = plt.subplots()
axs.imshow(dem[0], vmin=1900, vmax=4000, cmap='terrain')
```
